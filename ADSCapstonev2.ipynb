```python
import pandas as pd
import numpy as np
import folium
```

```python
# Read CSV file
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')

# Remove irrelevant tables
df = df[0]

# Drop NA values
df = df.dropna()

# Change "Postal code" to "Postal Code"
df.rename(columns = {'Postal code': 'Postal Code'}, inplace = True)

# Import Coordinates CSV
geocoords = 'Geospatial_Coordinates.csv'

# Create data frame for Coordinates
df2 = pd.read_csv(geocoords)

# Merge Neighborhoods Data Frame and Coordinates Data Frame
df = df.merge(df2, how="outer", on = 'Postal Code')
df
```

```python
# Create map
mooseWorldMap = folium.Map(location = [43.628841, -79.520999])

# Create postal code markers
for (index, row) in df.iterrows():
    folium.Marker(location = [row.loc['Latitude'], row.loc['Longitude']],
                 popup = row.loc['Postal Code']).add_to(mooseWorldMap)

mooseWorldMap
```
